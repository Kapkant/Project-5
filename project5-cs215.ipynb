```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
```

```python
energyAccess = pd.read_csv('number-of-people-with-and-without-electricity-access.csv')
#fixing columns by making names shorter
energyAccess.columns = ['Country', 'Code', 'Year', 'Access', 'No Access']
print(energyAccess)
```

Let's look at 4 different countries: Burundi, Nicaragua, Cambodia and Yemen

```python
burundi = energyAccess[energyAccess['Country']== 'Burundi']
nicaragua = energyAccess[energyAccess['Country']== 'Nicaragua']
cambodia = energyAccess[energyAccess['Country']== 'Cambodia']
yemen = energyAccess[energyAccess['Country']== 'Yemen']

print(burundi)
```

Now, the plot below shows the population of people without access to electricity, in millions, of 4 countries, recorded from 1997 to 2019.

```python
#setting plot size
fig, ax = plt.subplots(figsize=(14,20))

#plotting lines for each country
ax.plot(burundi['Year'], burundi['No Access'], color = 'red', marker = 'o')
ax.plot(nicaragua['Year'], nicaragua['No Access'], color = 'blue', marker = 'o')
ax.plot(cambodia['Year'], cambodia['No Access'], color = 'orange', marker = 'o')
ax.plot(yemen['Year'], yemen['No Access'], color = 'red', marker = 'o')

#setting axes, limits, ticks and labels
ax.axis([1997, 2019, 0, 9990000])

ax.grid(axis='y', linestyle='--')

ax.annotate('Burundi', xy=(2021, burundi['No Access'].iloc[-1]), xytext=(2022, 10000000), color='red', fontsize=15, arrowprops=dict(arrowstyle='-'))
ax.annotate('Nicaragua', xy=(2021, nicaragua['No Access'].iloc[-1]), xytext=(2022, 9500000), color='blue', fontsize=15, arrowprops=dict(arrowstyle='-'))
ax.annotate('Cambodia', xy=(2021, cambodia['No Access'].iloc[-1]), xytext=(2022, 9000000), color='orange', fontsize=15, arrowprops=dict(arrowstyle='-'))
ax.annotate('Yemen', xy=(2021, yemen['No Access'].iloc[-1]), xytext=(2022, 8500000), color='green', fontsize=15, arrowprops=dict(arrowstyle='-'))

ax.set_xticks([1997, 2005, 2010, 2019])
#fine tuning y-axis
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, pos: '{:.0f}M'.format(x/1000000) if x != 0 else '0'))
ax.set_yticks([0, 2000000, 4000000, 6000000, 8000000,9000000, 10000000, 11000000, 12000000])
# Set the y-axis tick label

plt.suptitle('Electricity Access, 1997 to 2021', fontsize =25)
plt.title('Measured in millions of people', fontsize = 20)
plt.xlabel('Years')
plt.ylabel('Population in millions')

plt.show()

```
